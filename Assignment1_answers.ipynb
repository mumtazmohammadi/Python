3.	Describe three different data types.

Answer:

Three different data types are:

•	Text type : String is a data type for a sequence of characters For example “Name”. String in python is denoted in single quote (‘ ’)or double quote (“ ”) or triple quote (‘’’)  example: “test name” 

•	Numeric data type : 
@Int : Integer or Int is a whole number positive or ngative without decimals of unlimited length eg 1, 3576687 , 598773988
@float : Float or Folating point number is a number positive or negative containing one or more decimals eg. 34.56, 78.23, 89e3
@complex : Complex numbers are written with “j” as the imaginary part eg. 3+5j, 5j


•	Sequence type:
@List : List are used to store multiple items in a single variable. List items are ordered, changeable, and allow duplicate values. List has strings, integers, Float or Boolean values. List items are inexed as – first item - [0] the second item is indexed [1] eg. [“list”, “name”,23,45.66,]

@tuple : List are used to store multiple items in a single variable. List items are ordered, unchangeable, and allow duplicate values. List has strings, integers, Float or Boolean values. List items are indexed as – first item - [0] the second item is indexed [1] eg. (“list”, “name”,23,45.66,)

•	Float: Float is a datatype used to store decimal numbers eg. 23.45 
•	Boolean: Boolean data type one of two values True or False. Its used to compare two values ‘True’ or ‘False’
